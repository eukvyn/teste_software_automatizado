```python
# CÓDIGO "GERAL" PARA TESTE DE SOFTWARE COM PYTHON E SELENIUM

#FAZENDO AS IMPORTAÇÕES NECESSÁRIAS
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep

#INICIANDO A AUTOMAÇÃO
#Abrindo o navegador e pesquisando pelo site 
nav = webdriver.Chrome()
nav.maximize_window()
nav.get(" URL DO SITE QUE SERÁ ACESSADO ")
sleep(2)

#Navegando até a tela de login/criação de conta
nav.find_element_by_xpath(' // ').send_keys(Keys.ENTER)
sleep(2)
nav.find_element_by_xpath('// ').send_keys(Keys.ENTER)
sleep(2)

#PROCESSO DE CRIAÇÃO DE CONTA
#Preenchendo campos 
nav.find_element_by_xpath(' // ').send_keys(" NOME ")
nav.find_element_by_xpath(' // ').send_keys(" EMAIL ")
nav.find_element_by_xpath(' // ').send_keys(" SENHA ")
nav.find_element_by_class_name(" // ").send_keys(Keys.SPACE) #MARCANDO CAIXA DE SELEÇÃO DE TERMOS DE USO
sleep(2)

#Avançando para o primeiro formulário de perfil
btnInscrever = nav.find_element_by_xpath(' // ')
ActionChains(nav).click(btnInscrever).perform()
sleep(10)

#PREENCHENDO OS CAMPOS DO PRIMEIRO FORMULÁRIO DE PERFIL
#Informando número de celular e Cidade
nav.find_element_by_xpath(' // ').send_keys(" TELEFONE ")
sleep(2)
nav.find_element_by_xpath(' // "]').send_keys(" CIDADE / UF ")
sleep(2)

#Informando Qual carreira está buscando - "QA/TESTES"
btnCarreira = nav.find_element_by_xpath(' // ')
ActionChains(nav).click(btnCarreira).perform()
sleep(2)

#Informando o Momento Profissional - " DISPONIVEL PARA SER CONTRATADO"
btnProfissional = nav.find_element_by_xpath(' // ')
ActionChains(nav).click(btnProfissional).perform()
sleep(2)

nav.find_element_by_xpath(' // ').send_keys(Keys.DOWN, Keys.ENTER)
sleep(2)

#Avançando para o próximo formulário
btnProximo = nav.find_element_by_xpath(' // ')
ActionChains(nav).click(btnProximo).perform()
sleep(5)

#PREENCHENDO OS CAMPOS DO SEGUNDO FORMULÁRIO DE PERFIL
#Informando a raça - "PARDO"
nav.find_element_by_xpath(' // ').send_keys(Keys.DOWN, Keys.DOWN, Keys.DOWN, Keys.DOWN, Keys.ENTER)
sleep(2)

#Informando a identidade de gênero - "HOMEM"
nav.find_element_by_xpath(' // ').send_keys(Keys.DOWN, Keys.ENTER)
sleep(2)

#Informando a orientação sexual - "HETEROSSEXUAL"
nav.find_element_by_xpath(' // ').send_keys(Keys.DOWN, Keys.ENTER)
sleep(2)

#Avançando para a próxima página
btnProximo2 = nav.find_element_by_xpath(' // ')
ActionChains(nav).click(btnProximo2).perform()
sleep(5)

#INFORMANDO QUE A PERSONALIZAÇÃO DO PERFIL FOI CONCLUIDA
btnConcluir = nav.find_element_by_xpath(' // ')
ActionChains(nav).click(btnConcluir).perform()
sleep(2)

#AVANÇANDO PARA O "DASHBOARD"
btnDashboard = nav.find_element_by_xpath(' // ')
ActionChains(nav).click(btnDashboard).perform()
sleep(2)

#AVANÇANDO PARA A TELA DE "CONSULTA"
btnVagas = nav.find_element_by_xpath(' // ')
ActionChains(nav).click(btnVagas).perform()
sleep(7)

#FAZENDO CONSULTA NA BARRA DE PESQUISA DO SITE "TESTE DE CONSULTA"
nav.find_element_by_xpath(' // ').send_keys("TESTE DE CONSULTA", Keys.ENTER)
```

```python
# CÓDIGO "GERAL" PARA TESTE DE SOFTWARE COM PYTHON E SELENIUM
# FAZENDO LOGIN COM O LINKEDIN

#FAZENDO AS IMPORTAÇÕES NECESSÁRIAS
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep

#INICIANDO A AUTOMAÇÃO
#Abrindo o navegador e pesquisando pelo site 
nav = webdriver.Chrome()
nav.maximize_window()
nav.get(" URL DO SITE QUE SERÁ ACESADO ")
sleep(2)

#Navegando até a tela de login/criação de conta
nav.find_element_by_xpath(' // ').send_keys(Keys.ENTER)
sleep(2)
nav.find_element_by_xpath(' // ').send_keys(Keys.ENTER)
sleep(2)

#Abrindo navegador para logar com o Linkedin
navLinkedin = webdriver.Chrome()
navLinkedin.maximize_window()
navLinkedin.get(" // ")

nav.quit()

#Fazendo Login na conta do Linkedin
navLinkedin.find_element_by_xpath(' // ').send_keys(" EMAIL DO LINKEDIN ")
sleep(2)

navLinkedin.find_element_by_xpath(' // ').send_keys(" SENHA DO LINKEDIN ", Keys.ENTER)
sleep(5)

#Após as credenciais do Linkedin serem informadas, é necessário navegar novamente pelo site e fazer login mais uma vez pelo Linkedin
btnPermitirLinkedin = navLinkedin.find_element_by_xpath(' // ')
ActionChains(navLinkedin).click(btnPermitirLinkedin).perform()
sleep(5)


import pyautogui # Importando biblioteca de Automação de mouse e teclado
# Fazer o mouse clicar na tela, de acordo com as cordenadas x e y passadas como parâmetro

btnLogin = navLinkedin.find_element_by_xpath(' // ')
ActionChains(navLinkedin).click(btnLogin).perform()
sleep(5)

btnCriarConta = navLinkedin.find_element_by_xpath(' // ')
ActionChains(navLinkedin).click(btnCriarConta).perform()
sleep(5)

btnLoginLinkedin = navLinkedin.find_element_by_xpath(' // ')
ActionChains(navLinkedin).click(btnLoginLinkedin).perform()
sleep(5)

pyautogui.click('CORDENADA X, CORDENADA Y, clicks=1') # CLICANDO PARA PERMITIR LOGIN COM LINKEDIN
```
